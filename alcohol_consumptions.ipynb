```python
import pandas as pd
import requests
import io
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
df
```

```python
df.groupby("continent")["beer_servings"].mean()
```

```python
df.groupby("continent")["wine_servings"].describe()
```

```python
df.groupby("continent")[["total_litres_of_pure_alcohol", "beer_servings",	"spirit_servings",	"wine_servings" ]].mean()
```

```python
df.groupby("continent")[["beer_servings",	"spirit_servings",	"wine_servings",	"total_litres_of_pure_alcohol"]].median()
```

```python
# using the agg method we can list the functions we want to apply to a particular series or dataframe
pd.DataFrame(df["spirit_servings"].agg(['mean', 'min', 'max']))
```
