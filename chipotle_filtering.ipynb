```python
import pandas as pd
import requests
import io
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')), sep="\t")
```

```python
df
```

```python
df.info()
```

```python
updated_price = [price.strip('$')for price in df['item_price']]
```

```python
final_price = pd.to_numeric(updated_price, errors="coerce")
```

```python
df['item_price'] = final_price
```

```python
# 1130 items have price greater than $10
df[df['item_price'] > 10]
```

```python
df[['item_name', 'choice_description', 'item_price']]
```

```python
df.sort_values('item_name', ascending=True)
```

```python
# the quantitiy for most expensive item ordered is 15
df[df['item_price'] == df['item_price'].max()]
```

```python
# veggie salad bowls was ordered 15 times
df[df['item_name'] == 'Veggie Salad Bowl']['item_name'].value_counts()
```

```python
# canned sode more than one can was ordered 20 times
df[(df['item_name'] == 'Canned Soda') & (df['quantity'] > 1)].info()
```
